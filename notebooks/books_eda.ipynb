I want to obtain the # of books with an average rating of > 4. 
Since average rating is a non-null object (see the next block for examples),
I am going to define a function that converts the 'number objects' to floats 
and the non-numeric objects to 'False'.

```python
# These are not numerics that can be typecasted 

#Unable to parse string " Jr.-Sam B. Warner" at position 4010
#Jr. Sam. Warner is in the position for average rating

#Unable to parse string " one of the founding members of this 
#.  Tolkien website)-Verlyn Flieger-Turgon (=David E. Smith)" at position 5685
#" one of the founding members of this #.  Tolkien website)-Verlyn Flieger-Turgon (=David E. Smith)" 
# is in the position for average rating
```

```python
def drop_function(x):
    """Function takes a cell of the dataframe and either
       a. converts the numeric object to a float, or
       b. changes the string to False"""
    try:
        return float(x)
    except ValueError:
        return False
```

```python
df['average_rating_number'] = df['average_rating'].apply(lambda x: drop_function(x));
```

```python
df.sample(5)
```

```python
print(f'This method dropped {13719 - len(df)} entries from the dataframe.')
```

```python
above_four_stars = df[df['average_rating_number'] > 4]
print(f'There are {len(above_four_stars)} books with more than 4 stars.')
```
